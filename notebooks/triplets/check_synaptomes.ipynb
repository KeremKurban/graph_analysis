```python
from scipy import sparse
import numpy as np
```

```python
target = 'cylinder300'
target_adj_with_proj = f'/gpfs/bbp.cscs.ch/project/proj112/circuits/CA1/20211110-BioM/data/ca1_ca3_to_{target}_synaptome.npz'
```

```python
adj = sparse.load_npz(target_adj_with_proj)
```

```python
local_adj = adj[:adj.shape[1]]
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
ax = plt.figure().add_axes([0.1, 0.1, 0.8, 0.8])
vmin = np.min(local_adj.A) + 0.01   # +0.01 to avoid log(0) inside the plot
vmax = np.max(local_adj.A)
ax.imshow(local_adj.A, cmap='Reds', norm=LogNorm(vmin=vmin, vmax=vmax))
```

```python
sampled_adj = adj[:adj.shape[1]*2,:]
```

```python
ax = plt.figure().add_axes([0.1, 0.1, 0.8, 0.8])
vmin = np.min(sampled_adj.A) + 0.01   # +0.01 to avoid log(0) inside the plot
vmax = np.max(sampled_adj.A)
ax.imshow(sampled_adj.A, cmap='Reds', norm=LogNorm(vmin=vmin, vmax=vmax))
plt.title('CA1 synaptome (cylinder)')
plt.savefig('../../output/synaptome_cylinder_with_proj.png', dpi=300)
```
