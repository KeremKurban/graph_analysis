# Synaptic weight extraction

We can easily get connectome or synaptome but these usually dont include spatial information or gsyn like synaptic parameters. We need to find a metric to get a weight between two nodes using all the circuit building steps.

Let neuron A and B a biophysically detailed model. The strength of the connection between the two depends on:

1. Number of synapses per connection
2. Synapse distances to soma
3. Conductance parameter
3. Whether synapses are clustered or separated. 
4. Whether they are on hot spots


----
The weighted adjacency matrix can be used to represent various properties of a graph, such as conductance, synapse distance to soma, and other metrics. [0] The adjacency matrix is a square matrix used to represent a finite graph, and the elements of the matrix indicate whether pairs of vertices are adjacent or not in the graph. [1] A weighted adjacency matrix includes a weight for each edge in the graph, which can be used to represent the strength or importance of the connection between two vertices. [2]

To normalize a weighted adjacency matrix for use in a graph convolutional network (GCN), there are two main approaches: [3]

Pick a suitable self-connection weight based on the meanings of the edges or normalize the graph beforehand. For example, in a correlation graph, where the edge strength measures the correlation between different nodes, the weight 1 would make sense as a node's correlation with itself is 1.
Introduce a new hyperparameter (HP) λ, and reformulate the normalizing procedure as Ã=A+λIN. The value of λ can be chosen based on the specific graph and the desired normalization.
It is important to note that the choice of normalization can affect the performance of the GCN, so it is recommended to experiment with different normalization techniques to find the best one for a given graph. [3]
