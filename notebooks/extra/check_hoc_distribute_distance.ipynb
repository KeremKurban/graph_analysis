```python
import numpy as np
import matplotlib.pyplot as plt

# Function to define the distance functions
def distance_function(x, param_expr):
    return eval(param_expr.replace("%.17g", str(x)))
```

```python

# List of section types and corresponding parameter expressions
section_types_params = {
    "apical": {
        "e_pas": "(-70.933702273152818-5*%.17g/150)",
        "ghdbar_hd": "(1. + 3./100. * %.17g)*1.9332485328531453e-05",
        "gkabar_kad": "(15./(1. + np.exp((300-%.17g)/50)))*0.0082032573537367138"
    },
    "basal": {
        "ghdbar_hd": "(1. + 3./100. * %.17g)*1.9332485328531453e-05",
        "e_pas": "(-72.646690840228288-5*%.17g/150)",
        "gkabar_kad": "(15./(1. + np.exp((300-%.17g)/50)))*0.0082032573537367138"
    },
    "somatic": {
        "ghdbar_hd": "(1. + 3./100. * %.17g)*1.9332485328531453e-05",
        "e_pas": "(-67.177143647261175-5*%.17g/150)"
    }
}

# Define the range of distance values
distances = np.linspace(0, 150, 100)

# Plot the distance functions for each section type and channel parameter
for section_type, params in section_types_params.items():
    plt.figure()
    plt.title(f"Distance functions for {section_type} section type")
    for param_name, param_expr in params.items():
        param_values = [distance_function(x, param_expr) for x in distances]
        plt.plot(distances, param_values, label=param_name)
    plt.xlabel("Distance from Soma (µm)")
    plt.ylabel("Parameter Value")
    plt.legend()
    plt.grid()
    plt.show()
```
