# extract node features

```python
circ_path = '/gpfs/bbp.cscs.ch/project/proj112/circuits/CA1/20211110-BioM/CircuitConfig'
slice_thickness = 400 # um
usertarget_path = f'slices{slice_thickness}.target'
circuit = Circuit(circ_path)

ATLAS_DIR = Path("/gpfs/bbp.cscs.ch/project/proj112/entities/atlas/20211004_BioM/")
CIRCUIT_DIR = Path('/gpfs/bbp.cscs.ch/project/proj112/circuits/CA1/20211110-BioM/')
c = Circuit((CIRCUIT_DIR / 'CircuitConfig').as_posix())
atlas = Atlas.open(ATLAS_DIR.as_posix())
orientation = atlas.load_data("orientation")
```

```python
NODE_FEATURES = ['etype', 'layer', 'model_template',
       'morph_class', 'mtype', 'region', 'synapse_class', 'x', 'y',
       'z', '@dynamics:holding_current', '@dynamics:threshold_current']
```

```python
node_features = c.cells.get().loc[:,NODE_FEATURES]
```

```python
node_features_preprocessed =  pd.get_dummies(node_features,columns=['etype', 'layer', 'model_template','morph_class', 'mtype', 'region', 'synapse_class'])
```

```python
node_features_preprocessed
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create an instance of MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler to your position data
scaler.fit(node_features_preprocessed[['x','y','z']])

# Transform and normalize the position data
node_features_preprocessed[['norm_x','norm_y','norm_z']] = scaler.transform(node_features_preprocessed[['x','y','z']])
```

```python
node_features_preprocessed = node_features_preprocessed.drop(['x','y','z'],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create an instance of MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler to your position data
scaler.fit(node_features_preprocessed[['x','y','z']])

# Transform and normalize the position data
node_features_preprocessed[['norm_x','norm_y','norm_z']] = scaler.transform(node_features_preprocessed[['x','y','z']])
```

```python
from sklearn.preprocessing import StandardScaler

# Create an instance of StandardScaler
scaler_holding = StandardScaler()

# Fit the scaler to your 'holding current' data
scaler_holding.fit(node_features_preprocessed[['@dynamics:holding_current']])

# Transform and normalize the 'holding current' data
node_features_preprocessed['holding_current_normalized'] = scaler_holding.transform(node_features_preprocessed[['@dynamics:holding_current']])
```

```python
from sklearn.preprocessing import StandardScaler

# Create an instance of StandardScaler
scaler_threshold = StandardScaler()

# Fit the scaler to your 'holding current' data
scaler_threshold.fit(node_features_preprocessed[['@dynamics:threshold_current']])

# Transform and normalize the 'holding current' data
node_features_preprocessed['threshold_current_normalized'] = scaler_threshold.transform(node_features_preprocessed[['@dynamics:threshold_current']])
```

```python
node_features_preprocessed = node_features_preprocessed.drop(['@dynamics:holding_current','@dynamics:threshold_current'],axis=1)
```
